```python
# Q024: Ras Protein and Cancer
show_question(
    title="Q024: Ras Mutation and Constitutive Kinase Activation",
    stars=6,
    year="2016 (and 2014, 2017, 2020, 2022)",
    marks=5,
    question_text="Explain how Ras-G12V mutation causes cancer. What is the mechanism of constitutive MAPK/PI3K activation?",
    answer_text="""<b>NORMAL RAS (Wild-type):</b>
- GDP-bound = INACTIVE (OFF)
- GTP-bound = ACTIVE (ON) — but transient!
- GAP (GTPase Activating Protein) converts GTP→GDP (signal termination)
- Result: ON/OFF cycling allows cell control

<b>MUTANT RAS (G12V = Glycine→Valine at position 12):</b>
- Codon 12 mutation disrupts GAP recognition site
- GAP CANNOT hydrolyze GTP!
- Ras LOCKED in GTP-bound (PERMANENTLY ON) ❌
- Consequence: Continuously recruits & activates Raf kinase

<b>DOWNSTREAM EFFECTS:</b>
1. Raf constantly phosphorylates MEK
2. MEK constantly phosphorylates ERK/MAPK
3. ERK enters nucleus → constant "DIVIDE" gene expression
4. Result: Uncontrolled cell division without external growth signal!

<b>PI3K ARM ALSO ACTIVATED:</b>
- Ras-GTP directly recruits PI3K (isoforms p110δ/γ)
- PI3K → PIP3 → AKT activation
- AKT → Phosphorylates mTOR, GSK3, FoxO
- Result: Pro-survival, anti-apoptotic signals ↑

<b>ROOT CAUSE OF CANCER:</b>
- ~30% of human cancers have RAS mutations (pancreatic 90%, colorectal 50%)
- Ras-G12V is the MOST COMMON (~90% of Ras mutations)
- Creates 'addiction' - cancer cell DEPENDS on this single mutation for survival
- Therapeutic target: Ras inhibitors (sotorasib) block GTP binding
"",
    mnemonic="G12V = GAP can't recognize = LOCKED in GTP = ALWAYS ON = ALWAYS DIVIDE",
    bn_term="রাস প্রোটিন (ক্ষুদ্র GTP বাঁধনকারী)"
)
```

```python
# Q040: TGF-β Signaling
show_question(
    title="Q040: TGF-β Canonical SMAD Pathway",
    stars=5,
    year="2014, 2015, 2017, 2018, 2020, 2022 (SIX TIMES!) ",
    marks=13,
    question_text="Draw and explain the TGF-β canonical pathway. Include SMAD-dependent and dual role in cancer.",
    answer_text="""<b>4-STEP CANONICAL PATHWAY (PSSR):</b>

P - Pathway Initiation:
- TGF-β dimer + Type II receptor (TβRII) = Constitutively active
- Forms heterotetrameric complex with TβRI (ALK5)

First S - SMAD Recruitment:
- TβRI phosphorylates SMAD2/3 (R-SMADs) on C-terminal tail
- Phosphorylation at Ser465/Ser467 creates binding site
- Creates MH2 domain activation

Second S - SMAD Complex:
- pSMAD2/3 + pSMAD4 (Co-SMAD with Smad4) form heterotrimer
- N-terminal MH1 domain binds DNA
- C-terminal MH2 domain recruits coactivators/corepressors

R - Response Gene Activation:
- Nuclear import (Importin-β1 + RanGTP)
- Binds SBE (SMAD Binding Elements) in promoters
- Cooperates with TCF/LEF transcription factors
- SMAD-TCF complexes activate CDK inhibitor genes:
  * p15 (INK4B) → blocks CDK4/6 activity
  * p21 (WAF1) → blocks CDK2 activity
- Result: Cell cycle arrest at G1/S checkpoint ✓

<b>DUAL ROLE IN CANCER:</b>

EARLY CANCER (Tumor Suppressor):
- p15/p21 ↑ → G1 arrest → prevents division ✓
- Apoptosis genes activated
- Epithelial differentiation promoted
- Inflammatory response suppressed via IL-10
- Result: **PROTECTION** against transformation

LATE CANCER (Tumor Promoter):
- SMAD4 deleted (~50% pancreatic cancer)
- OR SMAD7 upregulated (negative feedback)
- Non-canonical arms activated:
  * TAK1 (MKK3) → p38 MAPK → inflammation ↑
  * PAR1 → Rho → cytoskeletal changes
  * PI3K → AKT → pro-survival ↑
- Result: EMT (Epithelial→Mesenchymal transition) → invasion/metastasis ✓

<b>FLIP-FLOP MECHANISM (Key to understanding cancer):</b>
- Early: ~50% cells have p53, Rb functional → TGF-β = STOP signal
- Late: p53/Rb mutated (~70% cancers) → TGF-β loses control, activates survival
- This "context-dependence" is WHY TGF-β is paradoxical

<b>INHIBITORS & THERAPEUTICS:</b>
- ALK inhibitors (dorsomorphin, LY364947) block TβRI kinase
- SMAD3 inhibitors being developed
- Anti-TGF-β antibodies (ex: LY2157299) sequester ligand
- Strategy: Block early (protection) but carefully (side effects!)
"",
    mnemonic="TGF-β PSSR: Pathway → SMAD1 → SMAD2 → Response. Early=Tumor Suppressor, Late=Tumor Promoter!",
    bn_term="রূপান্তরকারী বৃদ্ধি ফ্যাক্টর বিটা পথ"
)
```

**Answer Format That Gets Full Marks:**

5. Mention disease application or clinical relevance

**Common Exam Mistakes (Avoid These!):**
- ❌ Forgetting that GAP inactivates Ras (saying Ras-GTP stays on forever)
- ❌ Confusing SMAD2/3 with SMAD4 roles (R-SMAD vs Co-SMAD)
- ❌ Missing the "3 cleavage steps" of Notch (TAN = TNF, ADAM, γ-secretase)
- ❌ Not explaining desensitization (GRK-β-arrestin mechanism)
- ❌ Forgetting TGF-β dual role context (early=suppressor, late=promoter)

**Last-Minute: Memorize These 5 Flows**
1. GPCR: Ligand → GPCR → Gαs-GDP*GTP → cAMP → PKA → Phosphorylase kinase → Glycogen breakdown
2. RTK: Growth factor → RTK phosphorylation → GRB2-SOS → Ras-GDP*GTP → Raf → MEK → ERK → Transcription
3. Wnt: Wnt+Frizzled → DVL → Axin relocates → GSK3 inactive → β-catenin stable → TCF/LEF → Gene ON
4. Notch: Delta → Notch → S1/(TNF) → S2/(ADAM) → S3/(γ-secretase) → NICD nuclear → CSL → HES genes
5. TGF-β: TGF-β → TβR1/R2 → SMAD2/3 phosphorylated → Complex with SMAD4 → SBE promoter → p15/p21 ↑ → G1 arrest
